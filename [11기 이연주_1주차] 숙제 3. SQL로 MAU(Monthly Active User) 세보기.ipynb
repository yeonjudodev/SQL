MAU 계산

```python
%%sql SELECT to_char(ts, 'YYYYMM') AS month, count(distinct userid) as user
FROM raw_data.user_session_channel A
LEFT JOIN raw_data.session_timestamp B ON A.sessionid = B.sessionid
GROUP BY month
ORDER BY 1;
```
